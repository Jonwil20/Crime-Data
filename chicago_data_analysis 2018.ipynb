```python
crime_pd3 = crime_pd2.loc[crime_pd2['Year'] == 2018, :]
#crime_pd3.head()
```

```python
grp_lat = crime_pd3.groupby(['Latitude'])
#grp_lat['Crime Type'].value_counts()
```

```python
map_crime =  folium.Map(location=[41.864073, -87.706819], zoom_start=11, tiles='CartoDB dark_matter')
crime_pd3['Latitude'] = crime_pd3['Latitude'].fillna(0)
crime_pd3['Longitude'] = crime_pd3['Longitude'].fillna(0)
#crime_pd3['Location'] = crime_pd3['Location'].astype(float)
for i in range(500):
    lat = crime_pd3['Latitude'].iloc[i]
    long = crime_pd3['Longitude'].iloc[i]
    popup_text= """" Community Index: {}<br>
                Block: {} <br>
                Crime Description: {} <br>"""
                
    popup_text = popup_text.format(i, crime_pd3['Block'][i], crime_pd3['Crime Type'][i])
    folium.CircleMarker(location=[lat,long], popup=popup_text, fill=True).add_to(map_crime)
    
map_crime
```

```python
map_crime.save('folium_map.html')
```

```python
df1 = crime_pd3['Latitude'].value_counts()
#df1
# create a dataframe from this 
x_df = pd.DataFrame({'Raw_string': df1.index,
                    'ValueCount': df1})
x_df.index = range(len(df1))
x_df.head()
```

```python
df2 = crime_pd3['Longitude'].value_counts()
x2_df = pd.DataFrame({'Raw_String2': df2.index,
                     'ValueCount2': df2})
x2_df.index = range(len(df2))
#x2_df.head
#df2
```

```python
# I am creating a map that give me the count of crime in chicago by location
map_crime2 = folium.Map(location=[41.864073, -87.706819], zoom_start=11, tiles='OpenStreetMap')
for i in range(500):
    lat= x_df['Raw_string'].iloc[i]
    long = x2_df['Raw_String2'].iloc[i]
    radius = float(x_df['ValueCount'].iloc[i] / 3)
    if x_df['ValueCount'].iloc[i] > 30:
        color = '#FF4500' # red
    else:
        color='#008080'# blue
    popup_text = """" Latitude: {}<br>
                Longitude: {} <br>
                Number of Crime: {} <br>"""
    popup_text = popup_text.format(lat,long,x_df['ValueCount'].iloc[i])
    folium.CircleMarker(location=[lat,long],popup=popup_text,radius=radius,color=color,fill=True).add_to(map_crime2)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
import numpy as np

weather_data = "weather_2018.csv"

weather_df = pd.read_csv(weather_data, parse_dates = ['dt_iso'],
                             index_col = ['dt_iso'] )
weather_df.head()
```

```python
reduced_weather_df = weather_df.loc[:, ["dt", "city_id",
                                    "temp", "temp_min", "temp_max", "pressure", "humidity", "wind_speed", "wind_deg", "weather_id", "weather_main", "weather_description", "weather_icon"]]
reduced_weather_df
```

```python
weather_2017 = "weather_2017.csv"
weather2_df = pd.read_csv(weather_data)

weather2_df.head()
```

```python
reduced_weather2_df = weather2_df.loc[:, ["dt", "city_id",
                                    "temp", "temp_min", "temp_max", "pressure", "humidity", "wind_speed", "wind_deg", "weather_id", "weather_main", "weather_description", "weather_icon"]]
reduced_weather2_df
```

```python
merge_weather = pd.merge(reduced_weather_df, reduced_weather2_df, on="dt")
merge_weather
```

```python
weather_master = pd.concat((reduced_weather_df,reduced_weather2_df),axis=0,join='outer',join_axes=None,ignore_index=False)
weather_master.head()
```

```python
organized_df = weather_master.rename(columns={"dt":"Date Timestamp","temp":"Temperature(Kelvin)", "temp_min":"Minimum Temperature(Kelvin)", "temp_max":"Maximum Temperature(Kelvin)"})
organized_df.head()
```

```python
temps_in_far = 1.8 * (organized_df["Temperature(Kelvin)"] - 273.15) + 32
temps_in_far2 = 1.8 * (organized_df["Minimum Temperature(Kelvin)"] - 273.15) + 32
temps_in_far3 = 1.8 * (organized_df["Maximum Temperature(Kelvin)"] - 273.15) + 32
organized_df["Temperature (Farenheight)"] = temps_in_far
organized_df["Minimum Temperature (Farenheight)"] = temps_in_far2
organized_df["Maximum Temperature (Farenheight)"] = temps_in_far3
organized_df.head()
```

```python
#Create a group based on the values in the 'maker' column
type_group = organized_df.groupby('weather_main')

# Count how many times each maker appears in our group
count_type = type_group['weather_main'].count()

count_type
```
